```python
def get_product(x:int, y:int) -> int:
    return x * y
```

```python
get_product(4, 5)
```

```python
get_product(4, "6")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_product(x:int, y:int) -> int:
    return x * y
```

```python
get_product(4, "6")
```
